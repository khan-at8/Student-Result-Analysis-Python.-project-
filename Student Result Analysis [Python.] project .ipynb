```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

#Drop unnmaed coloumn

```python
df = df.drop ("Unnamed: 0", axis = 1)
print(df.head())
```

#Gender distribution

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distri")
plt.show()
```

# from the above chart we have analysis that: the number of females in the data is more than the number of males.

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Students's Score ")
plt.show()
```

#from the above chart we have concluded that the ducation of the partents have a good impact on their scores.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent's Status and Students's Score ")
plt.show()
```

#from the above chart we have concluded that their is no/ngligble impact on the students score due to their parent'smarital status
